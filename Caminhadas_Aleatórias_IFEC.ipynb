(a) Escreva uma rotina para gerar caminhadas aleatórias de N passos em d  (=1 ou 2) dimensões, com cada passo uniformemente distribuído no intervalo (-1/2, 1/2) em cada dimensão. (Primeiro, gere os passos como um array N x d e, então, faça uma soma cumulativa.)

```python
def caminhada_aleatoria(N, d):
  passos = np.random.uniform(-1/5, 1/5, (N,d))
  caminhada = np.cumsum(passos, axis = 0)
  return caminhada
```

 Faça um gráfico de x_t por t para poucas caminhadas de 10.000 passos.
```python
n = 10000
d = 1
it = 10

x = np.arange(n)
plt.figure(figsize=(18,6))
for i in range(it):
  y = caminhada_aleatoria(n, d)
  plt.plot(x, y)
  plt.xlabel("t")
  plt.ylabel("x(t)")
plt.show()
```

É possível observar que, ao aumentar o número de passos em 100 vezes, a distância percorrida aumenta em cerca de 10 vezes. Por exemplo, com 10 passos, a distância fica perto de 0.4, enquanto com 1000 passos a distância fica perto de 4 e com 100000 passos a distância fica próxima de 40.

(b) Escreva uma rotina para determinar os pontos finais de W caminhadas aleatórias com N passos cada em d=2 dimensões. Faça um gráfico de dispersão das coordenadas finais de 10.000 caminhadas aleatórias com N=1 e 10, superpostos no mesmo gráfico. 

```python
def pontos_finais(W, N, d=2):
  finais = list()
  for i in range(W):
    caminhada = caminhada_aleatoria(N, d)
    final = caminhada[-1]
    finais.append(final.tolist())
  return finais

pontos_finais(5, 10)
```

```python
W = 10000
N_valores = [10, 1]
plt.figure(figsize=(8,8))
for n in N_valores:
  p_finais = pontos_finais(W, n)
  x = [z[0] for z in p_finais]
  y = [z[1] for z in p_finais]
  plt.scatter(x, y)
  plt.xlabel("x")
  plt.ylabel("y")
  plt.axis("square")
plt.show()
```
